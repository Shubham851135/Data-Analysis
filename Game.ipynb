```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read Csv file
df=pd.read_csv('odb.csv')
df
```

```python
df['HS'] = df['HS'].astype(str).str.replace('*', '', regex=False)
df['HS'] = df['HS'].astype(int)
df['6s'] = df['6s'].astype(str).str.replace('+', '', regex=False)
df['6s'] = df['6s'].astype(int)
df['4s'] = df['4s'].astype(str).str.replace('+', '', regex=False)
df['4s'] = df['4s'].astype(int)

```

```python
Highest_score=df[['Player','HS']].sort_values(by='HS',ascending=False).reset_index(drop= True).head(10)
print(Highest_score)
```

```python
plt.figure(figsize=(18,10))
bar=plt.bar(Highest_score['Player'],Highest_score['HS'])
plt.xlabel('Player')
plt.ylabel('HS')
plt.title('Highest_score')
plt.xticks(rotation=45)
plt.bar_label(bar)
```
