```python
read_csv['date_added']=pd.to_datetime(read_csv['date_added'],format='mixed')
```

```python
read_csv['date_added']
```

```python
cat_cols = ['type', 'rating']
for col in cat_cols:
    read_csv[col] = read_csv[col].astype('category')
```

```python
read_csv['country'] = read_csv['country'].fillna("Unknown")
read_csv['rating'] = read_csv['rating'].fillna(read_csv['rating'].mode()[0])
read_csv['duration'] = read_csv['duration'].fillna("Unknown")
```

```python
read_csv['country']
```

```python
read_csv['rating']
```

```python
read_csv['duration']
```

```python
read_csv[['duration_value', 'duration_unit']] = read_csv['duration'].str.extract(r'(\d+)\s*(\w+)')
```

```python
read_csv[['duration_value', 'duration_unit']]
```

```python
read_csv['duration_value'] = pd.to_numeric(read_csv['duration_value'], errors='coerce')
```

```python
read_csv['duration_value']
```

```python
df_country = read_csv.assign(
    country_exploded=read_csv['country'].str.split(',')
).explode('country_exploded')
# assign()+str.split(',') use to create a new column in which each row have a list of records('country')

df_country['country_exploded'] = df_country['country_exploded'].str.strip() # it is use to remove extra sapce

# explode() is use to divide each list into multiple rows
#like [india,usa] it will convert into india in 1st row and use in other row

```

```python
df_country
```

```python
df_genre = read_csv.assign(
    genre_exploded=read_csv['listed_in'].str.split(',')
).explode('genre_exploded')

df_genre['genre_exploded'] = df_genre['genre_exploded'].str.strip()
```

```python
df_genre
```

```python
df_cast=read_csv.assign(cast_exploded=read_csv['cast'].str.split(',')).explode('cast_exploded')
df_cast['cast_exploded']=df_cast['cast_exploded'].str.strip()
```

```python
df_cast
```

```python
read_csv['listed_in'] = read_csv['listed_in'].str.strip()
read_csv['country'] = read_csv['country'].str.strip()
read_csv['cast'] = read_csv['cast'].str.strip()
read_csv['director'] = read_csv['director'].str.strip()
```

```python
df_country = df_country.reset_index(drop=True)
df_genre = df_genre.reset_index(drop=True)
df_cast = df_cast.reset_index(drop=True)
```
